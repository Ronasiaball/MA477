<b>Time Series: Recreate the following plot</b>

(Hint: Convert the column `last_review` to `DateTime` format and reset it as the index of the dataframe)

```python
abnb['last_review']= pd.to_datetime(abnb['last_review']) #I got the code for these two cells from https://www.geeksforgeeks.org/convert-the-column-type-from-string-to-datetime-format-in-pandas-dataframe/. 4 Feb 2020.
abnb.sort_values('last_review', inplace=True)# this line from https://github.com/CoreyMSchafer/code_snippets/blob/master/Python/Matplotlib/08-TimeSeries/finished_code.py
```

```python
abnb.info()
```

```python
plt.style.use('seaborn')
Number_of_Reviews= abnb['number_of_reviews']
Price= abnb['price']
last_review= abnb['last_review']
plt.plot_date(last_review, Number_of_Reviews,linestyle='solid')

plt.gcf().autofmt_xdate()

plt.title('Fluctuations in Number of Reviews and Price over time ')
plt.xlabel('last_review')


plt.tight_layout()

plt.show()
#https://github.com/CoreyMSchafer/code_snippets/blob/master/Python/Matplotlib/08-TimeSeries/finished_code.py
```
